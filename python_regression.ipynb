```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
sns.set()
```

```python
boston = load_boston()
```

```python
df = boston.data
df = pd.DataFrame(df, columns=boston.feature_names)
```

```python
df["MEDV"] = boston.target
```

```python
df
```

```python
print(boston.DESCR)
```

```python
sns.heatmap(df.corr())
plt.show()
```

```python
df_y = df["MEDV"]
df_x = df[["RM"]]
```

```python
lr = LinearRegression()
```

```python
lr.fit(df_x,df_y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.intercept_ + lr.coef_*df_x
```

```python
sns.scatterplot(df_x["RM"], df_y)
plt.plot(df_x, lr.predict(df_x))
plt.show()
```
